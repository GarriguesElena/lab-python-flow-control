```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {
    "t-shirt": 15,
    "mug": 20,
    "hat": 10,
    "book": 4,
    "keychain": 13}

customer_orders = set()
```

```python
while True:
    product = input("Enter a product: ")

    if product in products:
        customer_orders.add(product)
    else:
        print("Product not available")
    
    another = input("Would you like to add another product? (yes/no)")

    if another == "yes":
        if another in products:
            customer_orders.add(product)
    else:
        break
    
```

```python
customer_orders
```

```python
for product in customer_orders:
    if product in inventory:
        inventory[product] -= 1

print(inventory)
```
